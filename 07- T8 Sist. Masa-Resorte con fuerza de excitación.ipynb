```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist
```

## Caso 1: $\omega / \omega_{n} < 1$

```python
#Variables del problema
wn = 22 #Frecuencia natural del sistema
w = 2 #Frecuencia de la fuerza de excitación
A = 1.0 #Constante de la respuesta transitoria
B = .4 #Constante de la respuesta en estado estable

#Arreglos utilizados
x = [] #Respuesta total del sistema
Y = [] #Arreglo auxiliar
cos1 = [] #Respuesta transitoria
cos2 = [] #Respuesta en estado estable

#Escala de tiempo
time = np.linspace(0, 5, 10000)

#Cálculo numérico de las funciones
for t in time:
    cos1.append(A*np.cos(wn*t))
    cos2.append(B*np.cos(w*t))
    Y.append(0)

x = [sum(c) for c in zip(cos1,cos2)]


#Graficación
fig, ax = plt.subplots(figsize=(10, 5), tight_layout=True)

#Modifaciones generales de los ejes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')


ax.set_xlim([0, 5])
ax.set_ylim([-3, 3])

ax.get_xaxis().set_ticks([])
plt.yticks([0], [r'$O$'])

ax.set_title("Respuesta total")
ax.set_ylabel(r"$X(t)$", fontsize=14)

#Mostrar gráficas
ax.plot(time, x, 'k', label='X(t)') 
ax.plot(time, cos2, 'k', linestyle='dashed', label=r'$X_{p}(t)$') #Respuesta Transitoria 
ax.text(4.4, -1.9, "Tiempo [s]", fontsize=14)
ax.annotate("", xy=(5, 0), xytext=(0, 0), arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=(0, 3), xytext=(0, -3), arrowprops=dict(arrowstyle="<->"))

#ax.plot(time, cos2, 'r') #Respuesta en estado estable
ax.legend(loc='lower left', fontsize=14)
plt.show()
```

## Caso 2: $\omega / \omega_{n} > 1$

```python
#Variables del problema
wn = 2 #Frecuencia natural del sistema
w = 12 #Frecuencia de la fuerza de excitación
A = 1.0 #Constante de la respuesta transitoria
B = .4 #Constante de la respuesta en estado estable

#Arreglos utilizados
x = [] #Respuesta total del sistema
Y = [] #Arreglo auxiliar
cos1 = [] #Respuesta transitoria
cos2 = [] #Respuesta en estado estable

#Escala de tiempo
time = np.linspace(0, 5, 10000)

#Cálculo numérico de las funciones
for t in time:
    cos1.append(A*np.cos(wn*t))
    cos2.append(-B*np.cos(w*t))
    Y.append(0)

x = [sum(c) for c in zip(cos1,cos2)]


#Graficación
fig, ax = plt.subplots(figsize=(10, 5), tight_layout=True)

#Modifaciones generales de los ejes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')


ax.set_xlim([0, 5])
ax.set_ylim([-3, 3])

ax.get_xaxis().set_ticks([])
plt.yticks([0], [r'$O$'])

ax.set_title("Respuesta total")
ax.set_ylabel(r"$X(t)$", fontsize=14)

#Mostrar gráficas
ax.plot(time, x, 'k', label='X(t)') 
ax.plot(time, cos1, 'k', linestyle='dashed', label=r'$X_{h}(t)$') #Respuesta en estado estable 
ax.text(4.4, -.3, "Tiempo [s]", fontsize=14)
ax.annotate("", xy=(5, 0), xytext=(0, 0), arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=(0, 3), xytext=(0, -3), arrowprops=dict(arrowstyle="<->"))

ax.legend(loc='lower left', fontsize=14)
plt.show()
```

## Caso 3: $\omega / \omega_{n} \approx 1$

```python
#Variables del problema
wn = 10 #Frecuencia natural del sistema
w = 10 #Frecuencia de la fuerza de excitación
A = 1.0 #Constante de la respuesta transitoria
B = .4 #Constante de la respuesta en estado estable

#Arreglos utilizados
x = [] #Respuesta total del sistema
Y = [] #Arreglo auxiliar
cos1 = [] #Respuesta transitoria
cos2 = [] #Respuesta en estado estable

#Escala de tiempo
time = np.linspace(0, 5, 10000)

#Cálculo numérico de las funciones
for t in time:
    cos1.append(A*np.cos(wn*t))
    cos2.append(0.75*t*np.sin(wn*t))
    Y.append(0)

x = [sum(c) for c in zip(cos1,cos2)]


#Graficación
fig, ax = plt.subplots(figsize=(10, 5), tight_layout=True)

#Modifaciones generales de los ejes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')


ax.set_xlim([0, 5])
ax.set_ylim([-5, 5])

ax.get_xaxis().set_ticks([])
plt.yticks([0], [r'$O$'])

ax.set_title("Respuesta total")
ax.set_ylabel(r"$X(t)$", fontsize=14)

#Mostrar gráficas
ax.plot(time, x, 'k', label='X(t)') 
ax.text(4.75, -.9, "t [s]", fontsize=14)
ax.annotate("", xy=(5, 0), xytext=(0, 0), arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=(0, 5), xytext=(0, -5), arrowprops=dict(arrowstyle="<->"))

ax.legend(loc='lower left', fontsize=14)
plt.show()
```
