```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 4*np.pi, 0.01)
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = np.sin(0.5*x)

fig1, ax1 = plt.subplots(figsize=(10, 8), tight_layout=True)
ax1.plot(x, y1, 'r')
ax1.plot(x, y2, 'b')
ax1.plot(x, y3, 'g')
ax1.set_title(r"Variando la frecuencia de $F=ASin(\omega t)$", fontsize=14)
ax1.set_ylabel("F", fontsize=14)
ax1.set_xlabel("t", fontsize=14)
ax1.legend((r"$\omega_{0}$",r"$\omega>\omega_{0}$", r"$\omega<\omega_{0}$")
           , shadow=False, loc=(0.05, 0.10), handlelength=1, fontsize=12)
plt.show()
```

```python
x = np.arange(0.0, 4*np.pi, 0.01)
y4 = np.sin(x)
y5 = 2*np.sin(x)
y6 = 0.5*np.sin(x)

fig2, ax2 = plt.subplots(figsize=(10, 8), tight_layout=True)
ax2.plot(x, y4, 'r')
ax2.plot(x, y5, 'b')
ax2.plot(x, y6, 'g')
ax2.set_title(r"Variando la amplitud de $F=ASin(\omega t)$", fontsize=14)
ax2.set_ylabel("F", fontsize=14)
ax2.set_xlabel("t", fontsize=14)
ax2.legend((r"$A_{0}$",r"$A>A_{0}$", r"$A<A_{0}$")
           , shadow=False, loc=(0.05, 0.10), handlelength=1, fontsize=12)
plt.show()
```

```python
x = np.arange(0.0, 4*np.pi, 0.01)
y7 = np.sin(x)
y8 = np.sin(x + np.pi/3)
y9 = np.sin(x - np.pi/3)

fig3, ax3 = plt.subplots(figsize=(10, 8), tight_layout=True)
ax3.plot(x, y7, 'r')
ax3.plot(x, y8, 'b')
ax3.plot(x, y9, 'g')
ax3.set_title(r"Variando la fase de $F=ASin(\omega t + \phi)$", fontsize=14)
ax3.set_ylabel("F", fontsize=14)
ax3.set_xlabel("t", fontsize=14)
ax3.legend((r"$\phi=0$",r"$\phi = \pi/3$", r"$\phi = -\pi/3$")
           , shadow=False, loc=(0.05, 0.10), handlelength=1, fontsize=12)
plt.show()
```
